```python
from easynmt import EasyNMT
```

```python
model=EasyNMT('opus-mt')
```

```python
sentences=['Jesus is alive',
           'Hands on projects is our priority',
           "I don't have anything",
           "It is  what is!:legend Balise said",
           "I don't have anything to give you baby" ,
           'good morning -how are you',
           'I LOVE YOU ',
           'THANK YOU !']
```

```python
translations=model.translate(sentences,target_lang='fr')
```
